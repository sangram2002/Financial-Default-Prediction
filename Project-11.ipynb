### Description:

### Data Dictionary

# Financial Metrics & Ratios

## Net Worth & Assets
- **Networth Next Year**: Net worth of the customer in the next year.
- **Total Assets**: Total assets of the customer.
- **Net Worth**: Net worth of the customer in the present year.
- **Total Income**: Total income of the customer.
- **Change in Stock**: Difference between the current value of the stock and the value of stock on the last trading day.
- **Total Expenses**: Total expenses incurred by the customer.

## Profit & Taxation
- **Profit After Tax (PAT)**: Profit after tax deduction.
- **PBDITA**: Profit before depreciation, income tax, and amortization.
- **PBT**: Profit before tax deduction.
- **Cash Profit**: Total cash profit.

## Profitability Ratios
- **PBDITA as % of Total Income**: `PBDITA / Total Income`
- **PBT as % of Total Income**: `PBT / Total Income`
- **PAT as % of Total Income**: `PAT / Total Income`
- **Cash Profit as % of Total Income**: `Cash Profit / Total Income`
- **PAT as % of Net Worth**: `PAT / Net Worth`

## Revenue & Income Sources
- **Sales**: Sales done by the customer.
- **Income from Financial Services**: Income generated from financial services.
- **Other Income**: Income from other sources.

## Capital & Equity
- **Total Capital**: Total capital of the customer.
- **Reserves & Funds**: Total reserves and funds available.
- **Borrowings**: Total amount borrowed by the customer.
- **Shareholders' Funds**: Amount of equity in a company belonging to shareholders.
- **Cumulative Retained Profits**: Total cumulative profit retained by the customer.

## Liabilities & Provisions
- **Current Liabilities & Provisions**: Current liabilities of the customer.
- **Deferred Tax Liability**: Future income tax the customer will pay due to the current transaction.
- **Contingent Liabilities**: Liabilities due to uncertain events.

## Financial Stability & Efficiency
- **Capital Employed**: `Current Assets - Current Liabilities`
- **TOL/TNW Ratio**: `Total Liabilities / Total Net Worth`
- **Total Term Liabilities / Tangible Net Worth**: `(Short-Term + Long-Term Liabilities) / Tangible Net Worth`
- **Contingent Liabilities / Net Worth (%)**: `Contingent Liabilities / Net Worth`

## Assets & Investments
- **Net Fixed Assets**: Purchase price of all fixed assets.
- **Investments**: Total invested amount.
- **Current Assets**: Assets expected to be converted to cash within a year.
- **Net Working Capital**: `Current Assets - Current Liabilities`

## Liquidity Ratios
- **Quick Ratio (times)**: `Total Cash / Current Liabilities`
- **Current Ratio (times)**: `Current Assets / Current Liabilities`
- **Debt to Equity Ratio (times)**: `Total Liabilities / Shareholders' Equity`
- **Cash to Current Liabilities (times)**: `Total Liquid Cash / Current Liabilities`
- **Cash to Average Cost of Sales per Day**: `Total Cash / Average Cost of Sales`

## Turnover Ratios
- **Creditors Turnover**: `Net Credit Purchase / Average Trade Creditors`
- **Debtors Turnover**: `Net Credit Sales / Average Accounts Receivable`
- **Finished Goods Turnover**: `Annual Sales / Average Inventory`
- **WIP Turnover**: `Cost of Goods Sold / Average Work-in-Progress Inventory`
- **Raw Material Turnover**: `Cost of Goods Sold / Average Raw Material Inventory`

## Stock & Market Metrics
- **Shares Outstanding**: `Issued Shares - Shares Held by the Company`
- **Equity Face Value**: Cost of equity at the time of issuance.
- **EPS (Earnings Per Share)**: `Net Income / Total Number of Outstanding Shares`
- **Adjusted EPS**: `Adjusted Net Earnings / Weighted Average Number of Common Shares Outstanding`
- **Total Liabilities**: Sum of all types of liabilities.
- **PE on BSE**: `Current Stock Price / Earnings Per Share`

```python
Default_X = Default.drop('default', axis = 1)
Default_Y = Default['default']
```

```python
def clip_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
for column in Default_X.columns:
    lr,ur=clip_outlier(Default[column])
    Default_X[column]=np.where(Default_X[column]>ur,ur,Default_X[column])
    Default_X[column]=np.where(Default_X[column]<lr,lr,Default_X[column])
```

```python
Default = pd.concat([Default_X, Default_Y], axis = 1)
```
